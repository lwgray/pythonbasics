````python
# Use these imports
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
````

### Show Wages by Age

````python
# Import Wages DataSet - Use Pandas - 'Wage.csv' data located in seminar3 folder 
data = pd.read_csv('Wage.csv')

# Define the X and Y values
x = data['age']
y = data['wage']
````

````python
# Plot scatter plot with gray points
sns.scatterplot(x,y, color='gray')
plt.show()
````

````python
#Extra credit: Plot with smoothed regression line and blue regression line
sns.regplot(x,y,lowess=True, scatter_kws={"color": "gray"}, line_kws={"color": "blue"})

# Extra credit (Change range of x ticks)
plt.xticks(np.arange(20, 81, 20.0))
plt.show()
````

### Show Wages by Year

````python
# Define X and Y
x = data['year']
y = data['wage']

# Plot scatter plot
sns.scatterplot(x,y, color='grey')
plt.show()

# Extra Credit Draw with trend line (Could use a different plot than used above)
sns.regplot(x,y, fit_reg=True, scatter_kws={"color": "gray"}, line_kws={"color": "blue"})
plt.xticks(np.arange(2002, 2011, 1.0))
plt.show()
````

### Show wages by Education Level

````python
# Define X and Y data
x = data['education']
y = data['wage']

# Extra Credit: create education levels that are mapped to numbers
x = data['education'].astype("category").cat.codes

# Plot boxplot
sns.boxplot(x,y, whiskerprops={"linestyle":'dashed'})

#Extra Credit: Plot with dashed whiskers
sns.boxplot(x,y, whiskerprops={"linestyle":'dashed'})

# Extra Extra Credit: Save plot as 'png'
plt.savefig('boxplot.png')
plt.show()
````
